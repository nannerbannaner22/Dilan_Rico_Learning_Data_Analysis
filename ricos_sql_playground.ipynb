# Grouping Animals by Risk Level and Extinction Status

This query is grouping animals based on two things:

- their **risk level**
- whether they are **extinct or not**

---

## What each part does

### 1. `FROM fun.animals a`

Start with the `animals` table.

- `a` is just a short nickname for `animals`

---

### 2. `JOIN fun.risklevel r`

Connect each animal to its risk level.

- `r` is a short nickname for `risklevel`

---

### 3. `ON a.RiskLevelId = r.RiskLevelID`

This tells SQL how to match the two tables together.

It means:

- take each animal
- find the matching row in `risklevel`
- match them using `RiskLevelId`

So if an animal has `RiskLevelId = 3`, SQL finds the row in `risklevel` where `RiskLevelId = 3`.

---

### 4. `SELECT count(a.AnimalId) as [Count], r.RiskName`

This tells SQL what to show in the result.

It returns:

- the **number of animals** in each group
- the **risk level name** for that group

Example output:

| Count | RiskName     |
|------:|--------------|
| 12    | Caution      |
| 8     | Spicy        |
| 5     | Harmless-ish |

---

## What `GROUP BY` means in simple terms

### `GROUP BY r.RiskName, IsExtinct`

This is the part that creates the groups.

It tells SQL:

> Put the rows into buckets based on `RiskName` and `IsExtinct`.

So instead of treating all animals as one big list, SQL splits them into separate groups such as:

- `Caution + extinct`
- `Caution + not extinct`
- `Spicy + extinct`
- `Spicy + not extinct`

Then this part:

`count(a.AnimalId)`

counts how many animals are in each bucket.

---

## Easy way to think about it

`GROUP BY` is like sorting animals into labeled boxes.

The labels here are:

- **risk level**
- **extinction status**

Then SQL counts how many animals are in each box.

---

## What `HAVING` is doing here

### `HAVING a.IsExtinct = 0`

This is trying to keep only the groups where `IsExtinct = 0`.

In your data:

- `0 = not extinct`
- `1 = extinct`

So this means:

> After grouping, only keep the groups for animals that are **not extinct**.

```python
select 
    count(a.AnimalId) as [Count], 
    r.RiskName 
from fun.animals a
join fun.risklevel r
on a.RiskLevelId = r.RiskLevelID
group by r.RiskName, IsExtinct
having a.IsExtinct = 0;
```
