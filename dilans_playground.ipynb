## MIN Function

```python
select
    MIN(a.WeightKg) as [Min]
from fun.Animals a where a.WeightKg >= 500;
```

## MAX Function

```python
select 
    MAX(a.WeightKg) as [Min]
from fun.Animals a where a.WeightKg >= 500;
```
