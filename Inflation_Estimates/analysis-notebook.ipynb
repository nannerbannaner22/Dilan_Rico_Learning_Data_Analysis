Price Inflation Estimates Table
The Price Inflation Estimates table contains monthly inflation estimates for various countries, covering the period from January 2007 to December 2026. Each record in the table represents the estimated inflation value for a specific country and month within this timeframe.

Key Features:

Timeframe: January 2007 – December 2026 (monthly increments)
Granularity: Country-level estimates
Purpose: To provide a comprehensive view of inflation trends across countries over time

```python
select top 10 * from dbo.PriceInflationEstimates pi;
```

GroupedPriceInflationEstimates View
The GroupedPriceInflationEstimates view provides a summarized, country-level aggregation of the data from the Price Inflation Estimates table. This view groups all records by country and ISO3 code, and then calculates key statistics for each group.

What this view does:

Groups the inflation estimate records by country and ISO3 code.
For each country group, it calculates:
Total Index: The count of records (rows) for that country.
Open: The average value of the OPEN column.
High: The average value of the High column.
Low: The average value of the Low column.
Close: The average value of the Close column.
Total: The sum of the Inflation column, rounded to two decimal places.
Intention:
This view is designed to provide a quick, high-level summary of inflation and related index values for each country. It enables analysts to easily compare inflation trends and index statistics across countries, supporting comparative economic analysis and reporting.

```python

select * from dbo.GroupedPriceInflationEstimates gpi;
```
